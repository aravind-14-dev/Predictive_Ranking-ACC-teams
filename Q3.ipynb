```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ACCGames1819.csv")
```

```python
data.head()
```

```python
away = data.iloc[:,2:18]
```

# Calculating away team's offensive efficiency for each game

```python
data['off_e_away'] = data.apply(lambda x: (x.AwayScore)/(x.AwayFGA - x.AwayORB + x.AwayTOV + (.475*x.AwayFTA)),axis=1)
```

```python
data['off_e_away'].head()
```

# Calculating away team's defensive efficiency for each game

```python
data['def_e_away'] = data.apply(lambda x:(x.HomeScore)/(x.HomeFGA - x.HomeORB + x.HomeTOV + (.475*x.HomeFTA)),axis=1)
```

```python
data['def_e_away'].head()
```

```python
data['net_away_efficiency'] = data['off_e_away'] - data['def_e_away']
```

```python
data['net_away_efficiency']
```

# Calculating home team's offensive and defensive efficiency for each game

```python
data['off_e_home'] = data.apply(lambda x: (x.HomeScore)/(x.HomeFGA - x.HomeORB + x.HomeTOV + (.475*x.HomeFTA)),axis=1)
```

```python
data['off_e_away'].head()
```

```python
data['def_e_home'] = data.apply(lambda x:(x.AwayScore)/(x.AwayFGA - x.AwayORB + x.AwayTOV + (.475*x.AwayFTA)),axis=1)
```

```python
data['def_e_home'].head()
```

```python
data['net_home_efficiency'] = data['off_e_home'] - data['def_e_home']
```

```python
data['net_home_efficiency']
```

```python
r_p = data[['AwayTeam','net_away_efficiency','HomeTeam','net_home_efficiency']]
```

```python
r_p
```

```python
apts=r_p.groupby(['AwayTeam']).sum()
apts=apts.drop(['net_home_efficiency'],axis=1)
```

```python
apts
```

```python
hpts=r_p.groupby(['HomeTeam']).sum()
hpts=hpts.drop(['net_away_efficiency'],axis=1)
```

```python
hpts
```

```python
hpts.reset_index(inplace=True)
apts.reset_index(inplace=True)
```

```python
hpts.head()
```

```python
overall = pd.concat([hpts,apts],axis=1)
```

```python
overall['Total_Efficiency'] = overall['net_home_efficiency'] + overall['net_away_efficiency']
```

```python
overall=overall.drop(columns=['net_home_efficiency','net_away_efficiency','AwayTeam'],axis=1)
```

```python
final = overall.sort_values(by=['Total_Efficiency'],ascending=False)
```

```python
final['Rank'] = final['Total_Efficiency'].rank(method='dense', ascending=False)
```

# Ranks based on overall efficiency

```python
final
```

```python
r = final['Rank']
final.drop(labels=['Rank'],axis=1,inplace=True)
final.insert(0,'Rank',r)
final
```

```python
final.reset_index(inplace=True)
```

```python
final.drop(labels=['index'],axis=1,inplace=True)
```
